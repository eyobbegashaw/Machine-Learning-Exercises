##### 1. Deliberately change the value of data to missing value in each feature of the dataset

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer, KNNImputer

 
data11 = pd.read_csv('iris copy.csv')
iris_df = data11.iloc[:, :-1]  

 
np.random.seed(42)

 
for column in iris_df.columns:
     
    iris_df.loc[iris_df.sample(n=10).index, column] = np.nan


print(iris_df.head())
```

##### 2. Impute the missing value data with mean value of the corresponding feature value 

```python
 
print("\nከመተካቱ በፊት የነበሩ የጎደሉ ዳታዎች ድምር:")
print(iris_df.isnull().sum())
 
for col in iris_df.columns:
    col_mean = iris_df[col].mean()
    
    iris_df[col] = iris_df[col].fillna(col_mean)

print("\nበ Mean ከተተካ በኋላ (First 5 rows):")
print(iris_df.head())

print("\nከተተካ በኋላ የነበሩ የጎደሉ ዳታዎች ድምር:")
print(iris_df.isnull().sum())

 
iris_df['Class'] = data11.iloc[:, 4]

print("Full Dataset with Imputed Values and Class:")
print(iris_df)
```

##### 3. Adjust the precision to 2 round off. 

```python
iris_df_rounded = iris_df.iloc[:, :-1].round(2) 

iris_df_rounded['Class'] = data11.iloc[:, 4]

print("ከጠጋጋ በኋላ (Rounded to 2 decimal places):")
print(iris_df_rounded.head())

data_combined = np.concatenate((iris_df_rounded, data11), axis=1)

print("\nየተቀላቀለው ዳታ (Combined Data - Numpy Array):")
print(data_combined)
```

##### 4. Impute the missing value data with most frequent value of the corresponding feature

```python
import pandas as pd
from sklearn.impute import SimpleImputer

 
data2 = pd.read_csv('iris copy.csv')

print("የመጨረሻዎቹ 5 ረድፎች (ከመተካቱ በፊት):")
print(data2.tail())
 
print("\nበየፊቸሩ የጎደሉ ዳታዎች ብዛት:")
print(data2.isnull().sum().sort_values(ascending=False))

 
imp_mode = SimpleImputer(strategy='most_frequent')

 
data2_imputed = pd.DataFrame(imp_mode.fit_transform(data2), columns=data2.columns)

print("\nበ Most Frequent ከተተካ በኋላ ያለው ዳታ (Numpy result):")
print(data2_imputed.values)  
```

##### 5. Impute the missing value data with a constant value 100 to every corresponding feature

```python
# 5. Impute with Constant 100
imp_constant = SimpleImputer(strategy='constant', fill_value=100)
data_constant = imp_constant.fit_transform(data2)

print("Data after Constant Imputation (Value=100):")
print(data_constant)
```

##### 6. Impute the missing value data with KNN where N=2

```python
from sklearn.impute import KNNImputer
 
knn_imputer = KNNImputer(n_neighbors=2)
 
numeric_data = data2.select_dtypes(include=[np.number])
data_knn = knn_imputer.fit_transform(numeric_data)

print("Data after KNN Imputation (N=2):")
print(data_knn)
```

##### 7. delete records of dataset with missing values  

```python
 
data_deleted = data2.dropna()

print("Data after dropping rows with missing values:")
print(data_deleted.head())
print("\nShape after dropping records:")
print(data_deleted.shape)
```

##### 8. Transform the dataset to Min-Max Normalization

```python
data_min_max = (data_deleted - data_deleted.min()) / (data_deleted.max() - data_deleted.min())
print("Min-Max Normalized Data:")
print(data_min_max.head())
```

##### 9. Transform the dataset to Z-Score Normalization

```python
data_zscore = (data_deleted - data_deleted.mean()) / data_deleted.std()
print("Z-Score Normalized Data:")
print(data_zscore.head())
```

##### 10. Take the age of patient from the hospital as follows  
Age={42, 15, 67, 55, 1, 29, 75, 89, 4, 10, 15, 38, 22, 77} 
Convert the given age value to the category of (‘Baby’, ‘Child’, ‘Teenage’, ‘Adult’, ‘Elderly’) suing 
binning method as cut value of [0, 3, 7, 17, 63, 99]

```python
import pandas as pd

# The given age data
age_data = [42, 15, 67, 55, 1, 29, 75, 89, 4, 10, 15, 38, 22, 77]

# Define the bin edges and the corresponding labels
# Bins: [0, 3], (3, 7], (7, 17], (17, 63], (63, 99]
bins = [0, 3, 7, 17, 63, 99]
labels = ['Baby', 'Child', 'Teenage', 'Adult', 'Elderly']

# Create a DataFrame
df_age = pd.DataFrame(age_data, columns=['Age'])

# Apply the binning method
df_age['Category'] = pd.cut(df_age['Age'], bins=bins, labels=labels)

print("Age Data with Categories:")
print(df_age)

# To see the counts in each category
print("\nCounts per Category:")
print(df_age['Category'].value_counts())
```

##### 11. Consider the following dataset and apply one hot encoding to the feature ‘city’ 

| City | Hawassa | Hawassa | Hawassa | Hawassa | Hawassa | Adama | Adama | Adama | Adama | BahirDar | BahirDar | BahirDar | BahirDar |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| **Area** | 260 | 300 | 320 | 360 | 400 | 260 | 280 | 330 | 360 | 260 | 290 | 310 | 360 |
| **Price** | 550,000 | 565,000 | 610,000 | 680,000 | 725,000 | 585,000 | 615,000 | 650,000 | 710,000 | 575,000 | 600,000 | 620,000 | 695,000 |

```python
import pandas as pd

# Creating the dataset based on your input
data = {
    'city': ['Hawassa', 'Hawassa', 'Hawassa', 'Hawassa', 'Hawassa', 
             'Adama', 'Adama', 'Adama', 'Adama', 
             'BahirDar', 'BahirDar', 'BahirDar', 'BahirDar'],
    'area': [260, 300, 320, 360, 400, 260, 280, 330, 360, 260, 290, 310, 360],
    'price': [550000, 565000, 610000, 680000, 725000, 585000, 615000, 
              650000, 710000, 575000, 600000, 620000, 695000]
}

df_city = pd.DataFrame(data)

# Applying One-Hot Encoding to the 'city' column
# We set dtype=int to get 1 and 0 instead of True/False
df_encoded = pd.get_dummies(df_city, columns=['city'], dtype=int)

print("Original DataFrame (First 5 rows):")
print(df_city.head())

print("\nDataFrame after One-Hot Encoding:")
print(df_encoded)
```
