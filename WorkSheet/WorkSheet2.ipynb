#### 1. Download the iris dataset from UCI machine learning repository and save it to your local machine

```python
import requests

# URL of the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Send a request to get the data
response = requests.get(url)

# Save the content to a local file named 'iris.csv'
with open("irisData.csv", "wb") as file:
    file.write(response.content)

print("Dataset downloaded and saved as irisData.csv")
```

```python
import pandas as pd
from sklearn.datasets import load_iris

 
iris = load_iris()

 
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

 
df['class'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
df = df.drop(columns=['target'])

 
df.to_csv('iris.csv', index=False)

print("Dataset downloaded and saved as iris.csv")
```
