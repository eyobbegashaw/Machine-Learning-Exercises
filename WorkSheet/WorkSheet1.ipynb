 <h4 style="color:green"><i> A.Generate line bar graph</i></h4>

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('student_data.csv')
 
plt.figure(figsize=(8, 5))
plt.bar(data['Dept'], data['Number of Students'], color='skyblue')

 
plt.xlabel('Departments')
plt.ylabel('Number of Students')
plt.title('Number of Students in Computing College')

 
plt.show()
```

 <h4 style="color:green"><i> B.Generate Pie chart </i></h4>

```python
import pandas as pd
import matplotlib.pyplot as plt

 
df = pd.read_csv('student_data.csv')
 
plt.figure(figsize=(8, 8))
plt.pie(df['Number of Students'], 
        labels=df['Dept'], 
        autopct='%1.2f%%',       
        startangle=140,          
        colors=['skyblue', 'lightgreen', 'orange', 'pink', 'cyan'])

 
plt.title('Percentage of Students in Computing College')

 
plt.axis('equal')  
plt.show()
```

  
 <h4 style="color:green"><i> C.Plot a bar chart by applying different color for each department</i></h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
 
df = pd.read_csv('student_data.csv')
 
my_colors = ['red', 'green', 'blue', 'orange', 'purple']
 
plt.figure(figsize=(8, 5))
bars = plt.bar(df['Dept'], df['Number of Students'], color=my_colors)
 
total = df['Number of Students'].sum()
for bar in bars:
    y_val = bar.get_height()
    percent = (y_val / total) * 100
    plt.text(bar.get_x() + bar.get_width()/2, y_val + 1, f'{percent:.2f}%', ha='center')

 
plt.xlabel('Departments')
plt.ylabel('Number of Students')
plt.title('Number of Students in Computing College')
 
plt.show()
```

 #### 3. Randomly generate a total of 1000 numbers  
* Plot histogram for the generated values by dividing to 20 bins 
* Title has to be ‘Histogram of randomly generated numbers’ 
* Xlabel has to be (‘Values) and ylabel has to be (‘Frequency’) 

```python
import numpy as np
import matplotlib.pyplot as plt

 
data = np.random.randn(1000)
 
plt.figure(figsize=(10, 6)) 
plt.hist(data, bins=20, color='teal', edgecolor='white', alpha=0.8)
 
plt.title('Histogram of randomly generated numbers', fontsize=14, fontweight='bold')
plt.xlabel('Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

 
plt.grid(axis='y', linestyle='--', alpha=0.5)

 
plt.tight_layout()
plt.show()
```

#### 4. Randomly generate 100 numbers with a mean distribution of 0 and standard deviation of 1 to 4 
* Generate box-plot graph 
* Title has to be ‘Box Plot’ 
* Xlabel has to be (‘Category’) and ylabel has to be (‘Values’)

```python
import numpy as np
import matplotlib.pyplot as plt

 
np.random.seed(42)  
sd1 = np.random.normal(0, 1, 100)
sd2 = np.random.normal(0, 2, 100)
sd3 = np.random.normal(0, 3, 100)
sd4 = np.random.normal(0, 4, 100)

data = [sd1, sd2, sd3, sd4]

 
plt.figure(figsize=(10, 6))

 
plt.boxplot(data, tick_labels=['SD=1', 'SD=2', 'SD=3', 'SD=4'], patch_artist=True,
            boxprops=dict(facecolor='lavender', color='darkblue'),
            medianprops=dict(color='red', linewidth=2))

 
plt.title('Box Plot', fontsize=14, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Values', fontsize=12)
 
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

#### 5. Randomly generate 100 numbers with a mean distribution of 0 and standard deviation of 1 to 6 
* Generate violinplot graph 
* Title has to be ‘Box Plot’, Xlabel has to be (‘Category’) and ylabel has to be (‘Values’) 


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 
np.random.seed(42)
data_list = []
for sd in range(1, 7):
    nums = np.random.normal(0, sd, 100)
    for n in nums:
        data_list.append({'Category': sd, 'Values': n})

df = pd.DataFrame(data_list)

 
plt.figure(figsize=(10, 6))

 
sns.violinplot(x='Category', y='Values', data=df, inner='box', 
               palette='viridis', hue='Category', legend=False)

 
plt.title('Box Plot', fontsize=15, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Values', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

#### 6.Consider the following data (Number of Instructors) 
 | Dept | IS | IT | CS | DS | SE |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **Instructor** | 15 | 13 | 18 | 10 | 12 |
* Generate radar chart 

```python
import plotly.graph_objects as go
import plotly.io as pio

# open on default browser
pio.renderers.default = "browser"

 
categories = ['IS', 'IT', 'CS', 'DS', 'SE']
instructors = [15, 13, 18, 10, 12]

categories_closed = categories + [categories[0]]
instructors_closed = instructors + [instructors[0]]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=instructors_closed,
    theta=categories_closed,
    fill='toself',
    line=dict(color='blue')
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 20])),
    showlegend=False,
    title="Number of Instructors per Department"
)

fig.show()
```

#### 7. Consider the following data  

| x | 1 | 2 | 3 | 4 | 5 | 6 | 7 | 8 |
| :--- | :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| **y1** | 10 | 15 | 25 | 30 | 35 | 45 | 50 | 55 |

*  Draw area chart  

```python
import matplotlib.pyplot as plt

 
x = [1, 2, 3, 4, 5, 6, 7, 8]
y1 = [10, 15, 25, 30, 35, 45, 50, 55]
y2 = [5, 10, 20, 25, 30, 35, 40, 50]

 
plt.figure(figsize=(10, 6))
 
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)

 
plt.title('Area Chart', fontsize=14, fontweight='bold')
plt.xlabel('X-Axis', fontsize=12)
plt.ylabel('Y-Axis', fontsize=12)

 
plt.grid(True, linestyle='--', alpha=0.5)

 
plt.show()
```

#### 8. Generate a number num from -2Xpi to 2*pi with  
* Generate 3 subplot  
* Plot1 num with sine of num 
* Plot2 num with cosine of num 
* Plot 3 num with tan of num

```python
import numpy as np
import matplotlib.pyplot as plt
 
num = np.linspace(-2 * np.pi, 2 * np.pi, 400)

 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# --- Plot 1: Sine ---
ax1.plot(num, np.sin(num), color='blue', linewidth=2)
ax1.set_title('Sine of num')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.axhline(0, color='black', lw=1) # የ X-ዘንግ መስመር

# --- Plot 2: Cosine ---
ax2.plot(num, np.cos(num), color='red', linewidth=2)
ax2.set_title('Cosine of num')
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.axhline(0, color='black', lw=1)

# --- Plot 3: Tan ---
 
tan_num = np.tan(num)
tan_num[np.abs(tan_num) > 10] = np.nan  
ax3.plot(num, tan_num, color='green', linewidth=2)
ax3.set_ylim(-10, 10)  
ax3.set_title('Tan of num')
ax3.grid(True, linestyle='--', alpha=0.7)
ax3.axhline(0, color='black', lw=1)

 
plt.suptitle('Trigonometric Subplots', fontsize=16, fontweight='bold')
plt.tight_layout()

 
plt.show()
```
